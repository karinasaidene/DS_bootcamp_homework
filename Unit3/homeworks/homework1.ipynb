```python
import pandas as pd
from sklearn.model_selection import train_test_split
from scipy import stats
import numpy as np
df=pd.read_csv("food-allergy-analysis-Zenodo.csv")
```

```python
df.head()
```

```python
#convert categorical data into numerical one
df=pd.get_dummies(df, columns=["GENDER_FACTOR", "RACE_FACTOR","ETHNICITY_FACTOR","PAYER_FACTOR","ATOPIC_MARCH_COHORT"], prefix=["GENDER", "RACE","ETHNICITY","PAYER","ATOPIC"])
```

```python
df.isna().sum()
```

```python
#dealing with missing values

df = df.fillna(df.mean())
np.where(df.values >= np.finfo(np.float64).max)
#the features TREENUT_ALG_END contains only NAN values so I am going to delete it
df.pop('TREENUT_ALG_END')
df.isnull().sum()











```

```python
df.pop('RACE_R2 - Asian or Pacific Islander')
```

```python
df
```

```python
#dealing with outliers
#detect outliers
df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]

```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1:].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
sk_model=LogisticRegression()
```

```python
sk_model.fit(x_train, y_train)
```

```python
#define the get_accuracy function
def get_accuracy(y_predict,y_test):
    correctly_classified=0
    for count in range(np.size(y_predict)):
        if y_test[count]==y_predict[count]:
            correctly_classified=correctly_classified+1
    acc=(correctly_classified/len(y_predict))*100
    print("accuracy on testset={:.2f}".format(acc))
    return acc
 
```

```python
y_predict=sk_model.predict(x_test)
```

```python

accur=get_accuracy(y_predict,y_test)
```
