```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df= pd. read_csv("clustering_data.csv")
df.head()
```

```python
df.info()
```

```python
X = df.values[:,0:]
X = np.nan_to_num(X)
```

```python

#executing the elbow method with th KMeans
distorsions = []
for k in range (2,20):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(X)
    distorsions.append(kmeans.inertia_)
    
fig = plt.figure(figsize = (15 ,5))
plt .plot (range(2,20),distorsions)
plt.grid(True)
plt.title("show curve")
```

from the elbow method we can se that the value that fits the most is 6.0

```python
random_state=1
y_pred = KMeans(n_clusters = 6, init ="random" ,random_state = random_state).fit_predict(X)
plt.scatter(X[:,0],X[:,1],c = y_pred)
```
